```python
import pandas as pd
import numpy as np
```

```python
master = pd.read_csv('datasets/master.csv', index_col=0)
```

```python
master.head()
```

```python
master.isna().sum()
```

```python
def replace_na(df, row):
    for column in ['PRCP', 'SNOW', 'TMIN', 'TMAX', 'TAVG', 'SNOWDepth']:
        if np.isnan(row[column]):
            row[column] = df[(df['County']==row['County']) & (df['CDC Week'] == row['CDC Week'])][column].mean()
    return row
```

```python
new = master.apply(lambda x: replace_na(master, x), axis=1)
```

```python
new.head()
```

```python
new.isna().sum()
```

```python
new.to_csv('datasets/master.csv')
```

```python
def fill_snow(df, row):
    for column in ['SNOW', 'SNOWDepth']:
        if np.isnan(row[column]):
            row[column] = df[df['CDC Week'] == row['CDC Week']][column].mean()
    return row
```

```python
new2 = new.apply(lambda x: fill_snow(new, x), axis=1)
```
